```python
import time
import random
import json
from kafka import KafkaProducer
producer = KafkaProducer(
    bootstrap_servers='localhost:9092',
    value_serializer=lambda v: json.dumps(v).encode('utf-8')
)
stocks = ["ITC", "TITAN"]
def generate_stock_data(batch_size=10, delay=1):
    while True:
        for _ in range(batch_size):
            stock_data = {
                "Date": time.strftime('%Y-%m-%d %H:%M:%S'),
                "Symbol": random.choice(stocks),
                "Open": round(random.uniform(100, 2000), 2),
                "High": round(random.uniform(100, 2000), 2),
                "Low": round(random.uniform(100, 2000), 2),
                "Close": round(random.uniform(100, 2000), 2),
                "Volume": random.randint(1000, 5000),
                "VWAP": round(random.uniform(100, 2000), 2)
            }
            # Send stock data to Kafka
            producer.send('stock_topic', stock_data)
            print(f"Produced: {stock_data}")
        
        # Wait for a short duration before sending the next batch
        time.sleep(delay)

if __name__ == "__main__":
    generate_stock_data(batch_size=10, delay=1)
```
